```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 프레임 생성 (여기서는 예시로 임의의 데이터를 사용)
df = pd.read_csv('train.csv')

# 'conversation' 컬럼의 문자열을 '\n'으로 나누고 그 길이를 담는 리스트 생성
df['conversation_length'] = df['conversation'].apply(lambda x: len(x.split('\n')))

# 분포 그래프 그리기
plt.figure(figsize=(10, 6))
plt.hist(df['conversation_length'], bins=range(1, df['conversation_length'].max()+2), align='left', edgecolor='black')
plt.xlabel('Number of Lines')
plt.ylabel('Frequency')
plt.title('Distribution of Conversation Length')
plt.show()
```

```python
df = pd.read_csv('train.csv')
# 'conversation' 컬럼의 전체 텍스트 길이를 담는 리스트 생성
df['total_text_length'] = df['conversation'].apply(len)

# 분포 그래프 그리기
plt.figure(figsize=(10, 6))
plt.hist(df['total_text_length'], bins=20, edgecolor='black')  # bins 값을 조정하여 히스토그램의 구간 수를 조절
plt.xlabel('Total Text Length')
plt.ylabel('Frequency')
plt.title('Distribution of Total Text Length')
plt.show()
```
